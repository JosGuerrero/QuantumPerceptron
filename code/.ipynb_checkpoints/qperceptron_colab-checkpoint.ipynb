```python

import numpy as np
import matplotlib.pyplot as plt
import math
import warnings
import random


def number_to_pm1_string(number, n_qubits):
    # convert a number into a plus and minus 1 (pm1) econding string

    m = 2**n_qubits
    max_value = 2**m

    # Domain check
    if number < 0 or number >= max_value:
        raise ValueError(
            f"number must be in range [0, {max_value - 1}] "
            f"for n_qubits = {n_qubits} (m = {m})."
        )

    binary = format(number, f'0{m}b')
    amplitudes = [(-1)**int(bit) for bit in binary]

    return amplitudes



def pm1_to_image(pm1_array):

    length = len(pm1_array)

    # infer N from length = 2^N
    n_qubits = int(math.log2(length))

    if 2**n_qubits != length:
        raise ValueError("Input length must be a power of 2.")

    # warn if N is odd (not square)
    if n_qubits % 2 != 0:
        warnings.warn(
            f"n_qubits = {n_qubits} is odd. "
            "Cannot form a square image. "
            "Resulting reshape may fail."
        )

    size = int(np.sqrt(length))
    return np.array(pm1_array).reshape(size, size)


def generate_sampled_dataset(n_qubits, n_samples):
    m = 2**n_qubits
    max_value = 2**m

    dataset = []
    for _ in range(n_samples):
        number = random.randint(0, max_value - 1)
        dataset.append(number_to_pm1_string(number, n_qubits))

    return dataset


def plot_pm1_image(im_array):

    h, w = im_array.shape

    fig, ax = plt.subplots()
    ax.imshow(im_array, cmap="gray", vmin=-1, vmax=1)

    # Grid at pixel boundaries
    ax.set_xticks(np.arange(0.5, w, 1))
    ax.set_yticks(np.arange(0.5, h, 1))

    # Remove tick labels
    ax.set_xticklabels([])
    ax.set_yticklabels([])

    ax.grid(True, color='black', linestyle='-', linewidth=1)

    plt.show()
```

HSGS based algorithm

Given a target input vector , the unitary Ui can be obtained starting from initial HN gate (see Eq. (12)). Then, the algorithm takes a series of iterative steps22 that are described below. First, we check whether there is any component with only one qubit in state |1 (i.e. of the form |00100) requiring a 1 factor, in the representation of |i on the computational basis. If so, the corresponding single qubit Z gate is applied by targeting the only qubit in state |1. Notice that this might introduce additional 1 factors in front of states with more than one qubit in state |1. Then, for p = 2, , N, we consider the components of the computational basis with exactly p qubits in state |1. For each of them, an additional 1 sign is introduced in front of its current amplitude (if it is needed and it was not previously introduced) by applying the corresponding CpZ between the p qubits in state |1. Similarly, if an unwanted sign is already present due to a previous step, this can be easily removed by applying the same CpZ. Since CpZ acts non trivially only on the manifold with p or more qubits being in state |1, the signs of all the elements with a lower number of |1 components are left unchanged. As a consequence, when p=N all the signs are the desired ones. As in the sign flip case, Uw can be obtained by slightly modifying the sequence of gates that would be used to generate |w. Indeed, one can start by first performing the HSGS tailored according to the 1 factors in |w. Since all the gates involved in HSGS are the inverse of themselves and commute with each other, this step is equivalent to the unitary transformation bringing |w back to the equally balanced superposition of the computational basis states |0. The desired transformation Uw is finally completed by adding parallel HN and NOTN gates (see Eq. (14)).

```python
# Qiskit Methods

def simulate_sv(qc, n_shots):

    # 1) Create Backend

    sim = AerSimulator(method = 'statevector')

    # 1.1) Transpile
    qc = transpile(qc, sim)

    # 2) Sampler V2 instance
    sampler = SamplerV2(sim)

    # 3) Run the sampler
    job = sampler.run(qc, shots = n_shots)

    # 4) Retrive results
    results_sim  = job.result()

    return results_sim

def getResultsAerBackend(simResult):
    pub_result = simResult[0]
    #get counts
    countsList = []
    #classicalReg = pub_result.c

    nshots = pub_result.data.c.num_shots
    counts = pub_result.data.c.get_counts() # In case there's a 0 probability
    counts.setdefault('0', 0)
    counts.setdefault('1', 0)
    #print(counts)
    result = counts['1'] / nshots
    return result

def simulateAerBackend(qc,n_shots, backend):

    '''
    # The backend needs to be outside the excecution loop
    # get a real backend from the runtime service

    service = QiskitRuntimeService()
    backend = service.least_busy(operational = True, simulator = False, min_num_qubits = qc.num_qubits)
    backend = service.backend("ibm_brisbane")
    '''
    # 1) Generate the simulator
    backend_sim = AerSimulator.from_backend(backend)


    # 2) Create the pass manager
    pm = generate_preset_pass_manager(optimization_level=3, backend = backend_sim)


    # 4) Transform into an ISA circuit (Transpilation)

    isaCircuit = pm.run(qc)
    #print('isaCircuit:')
    #circuit_drawer(isaCircuit[0], output="text", style={"backgroundcolor": "#EEEEEE"})

    print('num of gates: ', len(isaCircuit[0].data))
    # 5) Create the sampler object
    sampler = SamplerV2(backend_sim)

    # 6) Create the job
    simjob = sampler.run(isaCircuit, shots = n_shots)
    simResult = simjob.result()
    #print(np.shape(job.result()))
    #simResult = tr.getResultsFromJobs(simjob)
    print(simResult)
    result = getResultsAerBackend(simResult)

    return result

def noisySimulation(qc, n_shots):


    #1) Define Fake Backend

    backend = FakeBrisbane()
    # Pre transpilation

    # 2) Create the pass manager
    pm = generate_preset_pass_manager(optimization_level=0, backend = backend)
    #pm = generate_preset_pass_manager(optimization_level = 0)

    # 4) Transform into an ISA circuit (Transpilation)
    print('Transpiling')
    isaCircuit = pm.run(qc)
    #print(isaCircuit)

    # 5) Create the sampler object
    sampler = SamplerV2(backend)

    # 6) Create the job
    job = sampler.run(isaCircuit, shots = n_shots)
    #print(np.shape(job.result()))

    # 7) Return the job's result
    return job.result()

def getRSV(qc):
    nqubits = qc.num_qubits - 1
    
    sv_full = Statevector(qc).data.real
    #print(sv_full)
    sv = sv_full[:2**(qc.num_qubits - 1)]
   
    
    
    sv_real = np.sign(sv).astype(int) # cancels normalization
   
    return sv_real
def getAmplitudes(qc):
    sv = Statevector(qc)
    sv_real = np.round(np.real(sv),5) # Rounding necessary and justified, we only care about the signs
    sv_real = np.sign(sv_real).astype(int) # cancels normalization
    binary = np.int8((-1)*((sv_real - 1)/ 2))
    decimal = np.where(np.flip(binary)==1)
    return (sv_real, binary, decimal)
```

```python
# Manual Hypergraph State generation
'''
Naive implementation of the transformations for a two qubit quantum perceptron
'''
def u0(qc):
    pass
    
def u1(qc):
    qc.cz(0,1)
def u2(qc):
    qc.z(1)
    u1(qc)
def u3(qc):
    qc.z(1)
def u4(qc):
    qc.z(0)
    qc.cz(0,1)
def u5(qc):
    qc.z(0)

def u6(qc):
    qc.z([1,0])
def u7(qc):
    u6(qc)
    u1(qc)
def u8(qc):
    u1(qc)
    qc.x([0,1])
def u9(qc):
    qc.cz(0,1)
    qc.x([0,1])
    qc.cz(0,1)
def u10(qc):
    u5(qc)
    qc.x([0,1])
def u11(qc):
    qc.cz(0,1)
    qc.x([0,1])
    qc.z(1)
def u12(qc):
    u3(qc)
    qc.x([0,1])
def u13(qc):
    qc.cz(0,1)
    qc.x([0,1])
    qc.z(0)
def u14(qc):
    u7(qc)
    qc.x([0,1])
def u15(qc):
    qc.z(1)
    qc.x([0,1])
    qc.z(1)


# Uw
def uw_0(qc):
    pass
    
def uw_1(qc):
    qc.cz(0,1)
def uw_2(qc):
    u1(qc)
    qc.z(1)

def uw_3(qc):
    qc.z(1)
def uw_4(qc):
    qc.cz(0,1)
    qc.z(0)

def uw_5(qc):
    qc.z(0)

def uw_6(qc):
    qc.z([1,0])
def uw_7(qc):
    u1(qc)
    u6(qc)

def uw_8(qc):
    qc.x([0,1])
    u1(qc)

def uw_9(qc):
    qc.cz(0,1)
    qc.x([0,1])
    qc.cz(0,1)

def uw_10(qc):
    qc.x([0,1])
    u5(qc)

def uw_11(qc):
    qc.z(1)
    qc.x([0,1])
    qc.cz(0,1)

def uw_12(qc):
    qc.x([0,1])
    u3(qc)

def uw_13(qc):
    qc.z(0)
    qc.x([0,1])
    qc.cz(0,1)

def uw_14(qc):
    qc.x([0,1])
    u7(qc)

def uw_15(qc):
    qc.z(1)
    qc.x([0,1])
    qc.z(1)
```

```python
def getUnitary(target_state):
  unit = np.diag(target_state, dtype = np.int8)

def ui(qc, inputVec):
    N = qc.num_qubits
    rows = 2**(N)
    target = np.arange(N).tolist()
    ancilla = 2**1

    uiMat = getUnitary(inputVec)
    uiMatAnc = np.kron(np.eye(ancilla, dtype = np.int8), uiMat)

    Ui = UnitaryGate(uiMatAnc)
    qc.append(Ui, target)



def uw(qc, weigthVec):
    N = qc.num_qubits
    rows = 2**(N)
    target = np.arange(N).tolist()
    ancilla = 2**1

    uwMat = getUnitary(weigthVec)
    uwMatAnc = np.kron(np.eye(ancilla, dtype = np.int8), uwMat)

    Uw = UnitaryGate(uwMatAnc)
    qc.append(Uw, target)

```

```python
from qiskit import QuantumCircuit

# Unitaries

UiGates = {
    0: u0, 1: u1,  2: u2,  3: u3,  4: u4,  5: u5, 6: u6,  7: u7,  8: u8,  9: u9, 10: u10, 11: u11, 12: u12, 
    13: u13, 14: u14, 15: u15
}

UwGates = {
    0:uw_0, 1: uw_1,  2: uw_2,  3: uw_3,  4: uw_4,  5: uw_5, 6: uw_6,  7: uw_7,  8: uw_8,  9: uw_9, 10: uw_10, 
    11: uw_11, 12: uw_12, 13: uw_13, 14: uw_14, 15: uw_15
}

k = 15
N = 4

# There's no need of cheking k = 0

# Cheching the encoding of Psi_i 
for k in range(0,16):
    print(f'{'*'*25} k = {k} {'*'*25}')
    # Circuit Generation
    q_reg = QuantumRegister(2, 'q')
    c_reg = ClassicalRegister(1, 'c')
    qc = QuantumCircuit(q_reg,c_reg)

    # Encoding
    qc.h([0,1]) # | + > ^ n
    UiGates[k](qc) # Calling the sequence of gates to encode the binary number
    sv_real, binary, decimal = getAmplitudes(qc) # Get the emplitudes from the state vector 
    print(qc) # Print the circuit 
    print(f'k = {k} : m|> = {sv_real} ---> {binary} ----> {np.sum(2**decimal[0])}') # Check visually if 
    # the encoding was succesful

```

```python
## Iterating over all 


def perceptron(ku, kw, N):
    comb = 2**(2**(N - 1))
    qlist = np.arange(0,N-1).tolist()
    # Circuit preparation
    q_reg = QuantumRegister(N-1, 'q')
    a_reg = QuantumRegister(1, 'a')
    c_reg = ClassicalRegister(1, 'c')
    qc = QuantumCircuit(q_reg,a_reg, c_reg)
    
    # Encoding 
    qc.h([0,1]) # | + > ^ n
    qc.barrier(q_reg)
    UiGates[ku](qc) # Apply ui to get | _w > 
    
    UwGates[kw](qc) # Now Uw 
    
    qc.h([0,1]) # To arrive to |  > another haddamard is required
    qc.x([0,1])
    qc.mcx(qlist, N-1)
    qc.measure(N-1, 0)
    #print(qc)
    
    # Simulation 
    
    jobres = simulate_sv([qc], 2**8)
    projection = getResultsAerBackend(jobres)

    return projection
```

```python
import numpy as np
import matplotlib.pyplot as plt

Nimg = 16

# Your function
def compute_z(ki, kw):
    return (ki == kw) * 1  # example

# Build matrix
Z = np.zeros((Nimg, Nimg))
for i in range(Nimg):
    for j in range(Nimg):
        Z[i, j] = perceptron(i, j, 3)

plt.figure(figsize=(6,5))
plt.imshow(Z, origin='lower', cmap='viridis')
plt.colorbar(label=r"$|\sum w_{ij}|^2$")
plt.xlabel(r"$k_w$")
plt.ylabel(r"$k_i$")
plt.title("Quantum Perceptron Values")
plt.show()
```
